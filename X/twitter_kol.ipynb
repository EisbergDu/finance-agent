```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
import time
import re

driver = uc.Chrome()

all_usernames = set()

# 爬取前10页
for page in range(1,2):
    url = f"https://coinlaunch.space/influencers/?channelId=3&page={page}"
    print(f"\n正在爬取第 {page} 页...")
    
    driver.get(url)
    time.sleep(2)
    
    # 方法1: 精确定位 - 只找 td 内 class="clean-button" 的链接
    try:
        links = driver.find_elements(By.CSS_SELECTOR, "td a.clean-button")
        
        page_usernames = set()
        for link in links:
            href = link.get_attribute('href')
            if href and '/influencers/' in href:
                match = re.search(r'/influencers/([^/]+)/?$', href)
                if match:
                    username = match.group(1)
                    # 过滤掉分类标签（twitter, youtube等）
                    if username not in ['twitter', 'youtube', 'telegram', 'tiktok', 'instagram', 'discord',
                                       'founder', 'bitcoin', 'crypto-expert', 'analyst', 'trading', 
                                       'signals', 'investments', 'nft', 'gaming', 'news', 'altcoin', 'token-sale']:
                        page_usernames.add(username)
        
        print(f"  本页找到 {len(page_usernames)} 个用户名")
        
        # 显示前5个
        for username in list(page_usernames)[:5]:
            print(f"    - {username}")
        
        all_usernames.update(page_usernames)
        print(f"  累计总数: {len(all_usernames)} 个")
        
    except Exception as e:
        print(f"  出错: {e}")

print(f"\n=== 总计: {len(all_usernames)} 个唯一用户名 ===")

# 保存
with open('twitter_usernames_clean.txt', 'w', encoding='utf-8') as f:
    for username in sorted(all_usernames):
        f.write(username + '\n')

print("\n已保存到 twitter_usernames_clean.txt")

# 显示所有用户名
print("\n完整列表:")
for username in sorted(all_usernames):
    print(username)

driver.quit()
```
